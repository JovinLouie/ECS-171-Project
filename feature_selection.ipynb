```python
cross_tables = {}

for col in cat_info_fixed_cols:
    counts = pd.crosstab(df[col], df["Target"])
    percents = pd.crosstab(df[col], df["Target"], normalize="index")
    table = counts.astype(str) + " (" + (percents * 100).round(1).astype(str) + "%)"
    cross_tables[col] = table

for col in cat_info_fixed_cols:
    print(cross_tables[col])
    print()
```
